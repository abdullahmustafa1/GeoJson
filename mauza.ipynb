```python
import os
import json
import pandas as pd

# Directory path containing JSON files
directory_path = "/Users/kms/Documents/GitHub/GeoJson/DataJSON/Mauzas/"

# List to store data from all JSON files
all_rows = []

# Iterate through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith(".json"):  # Only process JSON files
        file_path = os.path.join(directory_path, filename)
        
        # Read the JSON file
        with open(file_path, 'r') as file:
            data = json.load(file)

        # Extract the 'features' list
        features = data.get('features', [])

        # Extract 'attributes' and 'geometry' from each feature
        for feature in features:
            attributes = feature.get('attributes', {})
            geometry = feature.get('geometry', {})
            all_rows.append({
                **attributes,
                'geometry': geometry  # Add geometry data as a single column
            })

# Create a DataFrame from all JSON files
df = pd.DataFrame(all_rows)

# Display the DataFrame
print(df.head())  # For console display
```
